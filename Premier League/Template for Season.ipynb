```python
def playerlist(season):
    URL =[]
    for i in range (0,100): 
        url = "https://footballapi.pulselive.com/football/players?pageSize=30&compSeasons="+str(season)+"&altIds=true&page="+str(i)+"&type=player&id=-1&compSeasonId="+str(season)
        URL.append(url)


    json_list= []

    for url in URL:
        data = requests.get(url)
        assist = data.json()
        json_list.append(assist)

    with open('playerslist.json', 'w+') as f:
        json.dump(json_list, f, sort_keys=True, indent=4)
        
    with open('playerslist.json') as f:
        df2 = json.load(f)
        
    full_name_col = []
    first_name_col = []
    last_name_col = []
    national_team_col = []
    position_col = []
    player_id_col = []
    epl_id_col = []
    club_col = []
    dob_col = []
    red_card_col = []

    for i in range (len(df2)):
        page = df2[i]
        players = page['content']
        for j in range (len(players)):
            full_name = players[j]['name']['display']
            full_name_col.append(full_name)

            first_name = players[j]['name']['first']
            first_name_col.append(first_name)

            last_name = players[j]['name']['last']
            last_name_col.append(last_name)

            #national_team = players[j]['nationalTeam']['country']
            #national_team_col.append(national_team)

            position = players[j]['info']['positionInfo']
            position_col.append(position)

            player_id = players[j]['playerId']
            player_id_col.append(player_id)


            epl_id = players[j]['id']
            epl_id_col.append(epl_id)

            #club = players[j]['currentTeam']['name']
            #club_col.append(club)

            #dob = players[j]['birth']['date']['label']
            #dob_col.append(dob)
    playerlist_df = pd.DataFrame({
        'full_name_col': full_name_col,
        'first_name_col': first_name_col,
        'last_name_col': last_name_col,

        'position_col': position_col,
        'player_id_col': player_id_col,
        'epl_id_col' : epl_id_col,

    })
    
    return playerlist_df
```

```python
def goals(season):

    URL =[]
    for i in range (0,50): 
        url = "https://footballapi.pulselive.com/football/stats/ranked/players/goals?page="+str(i)+"&pageSize=20&compSeasons="+str(season)+"&comps=1&compCodeForActivePlayer=EN_PR&altIds=true"
        URL.append(url)

    json_list = []

    for url in URL:
        data = requests.get(url)
        goals = data.json()
        json_list.append(goals)

    with open('goals.json', 'w+') as f:
        json.dump(json_list, f, sort_keys=True, indent=4)
    
    with open('goals.json') as f:
        df3 = json.load(f)
        epl_id_col = []
        value_col = []
        for i in range (len(df3)):
            page = df3[i]
            players = page['stats']['content']
            for j in range (len(players)):
                epl_id = players[j]['owner']['id']
                epl_id_col.append(epl_id)
                
                at_value = players[j]['value']
                value_col.append(at_value)
        goal_df = pd.DataFrame({'epl_id_col' : epl_id_col,
                            "goals": value_col,})
    return goal_df 
```
